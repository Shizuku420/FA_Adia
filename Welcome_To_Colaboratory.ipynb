```
import time

# Iterative approach with additional operations
def countdown_iterative(n):
    while n > 0:
        print(n)
        # Perform some additional operation
        _ = sum(range(n))
        # Print the sum of numbers up to n
        print("Sum of numbers up to", n, ":", _)
        n = n - 1
    print('Blastoff!')

# Test with different values of n
n_values = [100, 1000, 10000]

print("Testing iterative approach with additional operations:")
for n in n_values:
    start_time = time.time()
    countdown_iterative(n)
    end_time = time.time()
    print(f"Execution time for n={n}: {end_time - start_time} seconds")
    print("Blastoff!")
```
