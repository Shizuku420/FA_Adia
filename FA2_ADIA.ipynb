# **1. Write a Python function to multiply all the numbers in a list.**





```
def multiply_list(numbers):
    result = 1
    for num in numbers:
      result *= num
    return result

lists = (9,10,-2,4,7,3,-4,-5)
result = multiply_list(lists)
print(result)
```
